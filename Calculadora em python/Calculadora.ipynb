```python
def subtrair():
    while True:
        try:
            num1 = float(input('Insira o primeiro valor: '))
            while True:
                try:
                    num2 = float(input('Insira o segundo valor: '))
                    break
                except:
                    print("Você não digitou um número!")
        except:
            print("Você não digitou um número!")
        else:
            print('A subtração dos dois valores é:', num1 - num2)
            break
    
def soma():
    while True:
        try:
            num1 = float(input("insira o primeiro valor:"))
            while True:
                try:
                    num2 = float(input('Insira o segundo valor:'))
                    break
                except:
                    print("Você não digitou um número!")
        except:
            print("Você não digitou um numero!")
        else:
            print('A soma dos dois valores é :', num1 + num2)
            break
    
def multiplicar():
    while True:
        try:
            num1 = float(input('Digite o primeiro numero:'))
            while True:
                try:
                    num2 = float(input('Digite o segundo numero:'))
                    break
                except:
                    print("Você não digitou um número!")
        except:
            print("Você não digitou um número!")
        else:
            print('A multiplicação dos dois valores é:', num1 * num2)
            break
            
def dividir():
    while True:
        try:
            num1 = float(input('Digite o primeiro valor:'))
            while True:
                try:
                    num2 = float(input('Digite o segundo valor:'))
                    while num2 == 0:
                        num2 = float(input('Não é possível dividir por zero. Por favor, digite outro valor para o segundo número: '))
                    break
                except:
                    print("Você não digitou um número!")
        except:
                print('Você não digitou um número')
        else:
            print('A divisão dos dois valores é:', num1 / num2)
            break
    
continuacao = 's'
while continuacao == 's':   
    print("\n******************* Calculadora em Python *******************")
    print('\n')
    print('Escolha uma operação matemática:\n')
    print('1_Soma')
    print('2_Subtração')
    print('3_Multiplicação')
    print('4_Divisão\n')
    
    while True:
        try:
            opcao = int(input("Escolha uma das seguintes opções (1/2/3/4): "))
            while opcao not in [1,2,3,4]:
                print('Você não digitou a opção correta!')
                opcao = int(input("Escolha uma das seguintes opções (1/2/3/4): "))
        except:
            print("Você não digitou a opção correta!")
            continue
        else:
            if opcao == 1:
                soma()
                break
            elif opcao == 2:
                subtrair()
                break
            elif opcao == 3: 
                multiplicar()
                break
            elif opcao == 4:
                dividir()
                break
        
        
    continuacao = str(input('Deja continuar? [s/n]'))
    while continuacao != 's' and continuacao != 'n':
        continuacao = str(input('opção selecionada incorreta! Escolha [s/n]:'))
```
