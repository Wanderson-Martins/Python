```python
# Projeto 1 - Desaenvolvimento de game em Linguagem Python - Versão 1

#import
import random
from os import system, name

# Função para limpar a tela a cada execução
def limpa_tela():

	#Windows
	if name == 'nt':
		_= system('cls') #cls limpa a tela do windows. a função system retorna alguma coisa, coloca o _= para não retornar pq eu quero apenas a limpeza da tela

	#Mac ou Linux
	else:
		_system('clear') # clear limpa a tela de Mac e Linux

def fruta():
    listaPalavras = ["maçã", "banana", "laranja", "abacaxi", "uva", "morango", "pera", "melancia", "kiwi", "limão"]
    return random.choice(listaPalavras)

def pais():
    listaPalavras = ["brasil", "portugal", "china", "índia", "rússia", "indonésia", "paquistão", "nigéria", "bangladesh", "japão"]
    return random.choice(listaPalavras)

def nome():
    listaPalavras = ["maria", "joao", "ana", "pedro", "carla", "lucas", "juliana", "fernando", "cristiane", "gabriel"]
    return random.choice(listaPalavras)

def animal():
    listaPalavras = ["cachorro", "gato", "elefante", "leão", "tigre", "girafa", "macaco", "papagaio", "tartaruga", "peixe"]
    return random.choice(listaPalavras)

def planeta():
    listaPalavras = ["mercúrio", "venus", "terra", "marte", "júpiter", "saturno", "urano", "netuno", "plutão", "céres"]
    return random.choice(listaPalavras)

continuacao = 's'
while continuacao == 's':
    #Função
    def game():
        limpa_tela()

        print('******************* Bem vindo ao jogo da forca! *******************')
        print('\n')
        print("Temos os seguintes assuntos para jogar\n")
        print("1_ Frutas")
        print("2_ País")
        print("3_ Nome de Pessoas")
        print("4_ Animais")
        print("5_ Planetas\n")

        while True:
            try:
                opcao = int(input("Escolha um dos assuntos (1/2/3/4/5): "))
                while opcao not in [1,2,3,4,5]:
                    print("Você escolheu uma opção invalida!")
                    opcao = int(input("Escolha um dos assuntos (1/2/3/4/5): "))
            except:
                print("Você escolheu uma opção invalida!")
                continue
            else:
                if opcao == 1:
                    palavra = fruta()
                    break

                elif opcao == 2:
                    palavra = pais()
                    break

                elif opcao == 3:
                    palavra = nome()
                    break

                elif opcao == 4:
                    palavra = animal()
                    break

                elif opcao == 5:
                    palavra = planeta()
                    break

        #list comprehesion
        letraDescoberta = ['_' for letra in palavra]

        #numero de chances
        chances = 6

        #lista para letras erradas
        letrasErradas = []

        #loop enquanto número de chances for maior que zero
        while chances > 0:

            print(" ".join(letraDescoberta))
            print("\nChances restantes", chances)
            print("Letras erradas: ", " ".join(letrasErradas)) #O join junta oq esta no lado esquerto com oq esta no lado direito

            while True:
                #Tentativa
                tentativa = input("\nDigite uma letra: ").lower() 
                if not tentativa.isalpha():
                    print("Digite apenas letras.")
                    continue
                elif len(tentativa) != 1:
                    print("Digite apenas uma letra por vez.")
                else:
                    break

            #Condicional 
            if tentativa in palavra:
                index = 0
                for letra in palavra:
                    if tentativa == letra:
                        letraDescoberta[index] = letra 
                    index +=1
            else:
                chances -= 1
                letrasErradas.append(tentativa)

            if "_" not in letraDescoberta:
                print("\nParabéns! Você venceu, a palavra era:", palavra)
                break

            if chances == 0 and "_" in letraDescoberta:
                print("\nVocê perdeu, a palavra era:", palavra)

    #Bloco main
    if __name__=="__main__":
        game()
        continuacao = str(input("Deseja jogar de novo? [s/n]"))
        while continuacao not in ('s','n'):
            continuacao = str(input('opção incorreta, escolha s para jogar de novo ou n para encerrar o jogo'))
    
```
